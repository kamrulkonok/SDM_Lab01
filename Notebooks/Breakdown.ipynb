```python
import pandas as pd
```

```python
df_papers = pd.read_csv('../aryan_data/papers_info.csv')
df_papers.head()
```

```python
# Workshop, Conference & Journals

df_workshop = df_papers[df_papers['publicationTypes'] == 'Workshop'].copy()
df_workshop = df_workshop.drop(columns=['field_query', 'journal_volume', 'number_reviewers', 'reviewers', 'approved', 'comments', 'journal_name'])
df_workshop.head()
```

```python
df_conference = df_papers[df_papers['publicationTypes'] == 'Conference'].copy()
df_conference = df_conference.drop(columns=['field_query', 'journal_volume', 'number_reviewers', 'reviewers', 'approved', 'comments', 'journal_name'])
df_conference.head()
```

```python
df_journal = df_papers[df_papers['publicationTypes'] == 'JournalArticle'].copy()
df_journal = df_journal.drop(columns=['field_query', 'number_reviewers', 'reviewers', 'approved', 'comments', 'proceedings', 'edition'])
df_journal.head()
```

```python
df_workshop.to_csv('../usable_data/A1/workshop_info.csv', index=False)
df_conference.to_csv('../usable_data/A1/conference_info.csv', index=False)
df_journal.to_csv('../usable_data/A1/journal_info.csv', index=False)
```
