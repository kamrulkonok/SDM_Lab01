```python
import pandas as pd
```

```python
df_conference = pd.read_csv('../usable_data/A3/conference_info.csv')
df_conference.head()
```

To generate more data, I will change the `paperId`, `year`, `edition`

```python
len(df_conference)
```

```python
df_copy = df_conference.copy()
df_copy.info()
```

```python
df_copy['edition'] = df_copy['edition'] - 1
df_copy['year'] = df_copy['year'] - 1
df_copy['paperId'] = df_copy['paperId'] + 'a'

df_copy.head()
```

```python
# join both the dataframes together

df_conference = df_conference.append(df_copy, ignore_index=True)
df_conference.info()
```

```python
df_conference.tail()
```

```python
df_conference.to_csv('../usable_data/A3/conference_info.csv', index=False)
```
